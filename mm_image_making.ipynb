```python
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
from numpy.linalg import matrix_power
from scipy.linalg import expm, expm_frechet
import scipy
```

```python
# make a plot of three Gaussian signals, with two starting at 0 and one at pi/2

# number of steps
N = 1000
# number of trajectories
M = 3
# time
t = np.linspace(0, 10, N)
# covar 
covar = 0.00001*np.eye(3)

initial_vals = np.array([np.pi/2, 0, 0])
trajectories = np.array([initial_vals])

# generate the trajectories
for i in range(1, N):
    w = np.random.multivariate_normal(np.zeros(3), covar)
    new_vals = trajectories[-1] + w
    trajectories = np.append(trajectories, [new_vals], axis=0)
# plot on three different plots
fig, axs = plt.subplots(3, 1, figsize=(10, 10))
for m in range(M):
    axs[m].plot(t, trajectories[:, m])
    # increase the font size of labels and titles
    axs[m].tick_params(axis='both', which='major', labelsize=20)
    # increase the font size of labels and titles
    
    
    
    # plot lines at target vales
    axs[m].axhline(y=initial_vals[m], color='r', linestyle='--')
axs[0].set_ylabel('$\\theta_X$', fontsize=20)
axs[1].set_ylabel('$\\theta_Y$', fontsize=20)
axs[2].set_ylabel('$\\theta_Z$', fontsize=20)

axs[-1].set_xlabel('time [s]', fontsize=20)
# set a tight layout
plt.tight_layout()

   
```
